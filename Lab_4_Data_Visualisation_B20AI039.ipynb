# 1. Monthly changes in different food grain stocks

```python
df1.head()
```

```python
df1.info()
```

### Screen Display, Presentation and Poster Print View

```python
plt.figure(figsize = (18,14))
sns.set_style('darkgrid')

# plot 1    screen display
plt.subplot(2,2,1)
sns.set_context("notebook")

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o')

plt.legend( fontsize = '8')
plt.title('screen_display')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)

# plot 2    presentation
plt.subplot(2,2,2)
sns.set_context("talk")

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o' )

plt.legend( fontsize = '8')
plt.title('presentation')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)

# plot 3    poster print
plt.subplot(2,2,3)
sns.set_context("poster")

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o')

plt.title('Poster_print')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)

plt.legend( fontsize = '8')
plt.suptitle('Monthly Production of Foodgrains, Rice, Unmilled Paddy, and Wheat')
plt.tight_layout()
plt.show()
```

## Color Palletes

```python
# continuous color pallete
Continuous_color = sns.color_palette("flare", n_colors = 4)

one_color_sequential = sns.color_palette("dark:#5A9_r", n_colors = 4, as_cmap=False)

bipolar_sequential_color = sns.color_palette("blend:#7AB,#EDA", n_colors = 4, as_cmap=False)   # light yellow to dark red

custom_color = sns.color_palette(['#ff5a36', '#00968e', '#003f5c', '#ffa600'])        # given in the dv book
```

```python
plt.figure(figsize = (18,14))
sns.set_style('darkgrid')


# plot 1    Continuous_color
plt.subplot(2,2,1)

sns.set_context("talk")
# Set custom color palette
sns.set_palette(sns.color_palette(Continuous_color))

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o')

plt.legend( fontsize = '8')
plt.title('Continuous_color')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)



# plot 2    one_color_sequential
plt.subplot(2,2,2)

sns.set_context("talk")

sns.set_palette(sns.color_palette(one_color_sequential))

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o' )

plt.legend( fontsize = '8')
plt.title('one_color_sequential')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)



# plot 3    bipolar_sequential_color
plt.subplot(2,2,3)
sns.set_context("talk")

sns.set_palette(sns.color_palette(bipolar_sequential_color))

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o')

plt.title('bipolar_sequential_color')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)
plt.legend( fontsize = '8')



# plot 4    custom_color
plt.subplot(2,2,4)
sns.set_context("talk")

sns.set_palette(sns.color_palette(custom_color))

sns.lineplot(data=df1.melt(id_vars=['Particulars', 'Frequency', 'Unit'], var_name='Month', value_name='Stocks'), 
             x='Month', y='Stocks', hue='Particulars', marker = 'o')

plt.title('custom_color')
plt.xlabel('Month')
plt.ylabel('Production (million metric tons)')
plt.xticks(rotation = 40)
plt.legend( fontsize = '8')

plt.suptitle('Monthly Production of Foodgrains, Rice, Unmilled Paddy, and Wheat')
plt.tight_layout()
plt.show()
```

# 2. State wise Colleges of different Ayus Category.

```python
df2.head()
```

```python
df2 = pd.read_csv('MergerdAyush_Colleges_in_India_01-04-2021_0.csv', usecols=['State or Union Territory', 'Type of System', 'No. of Colleges - Govt', 'No. of Colleges - Non-Govt'])
df2['total_colleges'] = df2['No. of Colleges - Govt']  + df2 ['No. of Colleges - Non-Govt']


df2 = pd.pivot_table(df2, values='total_colleges', index='State or Union Territory', columns='Type of System')

df2.head()
```

# 3. Distribution of National, State and District roads.

```python
df3 = df3.loc[:, ['Name of the States', 'National Highways', 'State Highways', 'District Roads']]
df3.head()
```

```python
df3.info()
```

```python

# melt the dataframe to convert it to a grouped format by types of roads
df_melted = pd.melt(df3, id_vars=['Name of the States'], var_name='Road Type', value_name='Road Length')

# print the first few rows of the melted dataframe
print(df_melted)
```

```python
plt.figure(figsize = (20,12))
sns.set_context("talk")
# Create a bar chart of the road lengths for each state
sns.barplot(data=df_melted, x='Name of the States', y='Road Length', hue = 'Road Type')

plt.xticks(rotation=90)
plt.title('Distribution of National, State and District roads by State')
plt.show()
```

### Lograithmic scale

Logarithmic scale: The logarithmic scale compresses the data by increasing the distance between smaller values and decreasing the distance between larger values. This makes it easier to visualize data with a wide range of values or extreme outliers.

```python
plt.figure(figsize = (22,12))
sns.set_context("talk")
# Create a bar chart of the road lengths for each state
sns.barplot(data=df_melted, x='Name of the States', y='Road Length', hue = 'Road Type')

plt.xticks(rotation=90)
plt.yscale('log')
plt.ylabel('Road Length - Log Scale')
plt.title('Distribution of National, State and District roads by State - Logarithmic Scale')
plt.show()
```

### Screen Display, Presentation and Poster Print View

```python
plt.figure(figsize = (22,16))
sns.set_style('darkgrid')


# plot 1    
plt.subplot(2,2,1)

sns.set_context("talk")

# Create a bar chart of the road lengths for each state
sns.barplot(data=df_melted, x='Name of the States', y='Road Length', hue = 'Road Type')

plt.xticks(rotation=90)
plt.yscale('log')
plt.ylabel('Road Length - Log Scale')
plt.title('Preentation View')
plt.legend( fontsize = '10')

# plot 2    
plt.subplot(2,2,2)

sns.set_context("poster")

sns.barplot(data=df_melted, x='Name of the States', y='Road Length', hue = 'Road Type')

plt.xticks(rotation=90)
plt.yscale('log')
plt.ylabel('Road Length - Log Scale')
plt.title('Poster Print View')
plt.legend( fontsize = '10')


# plot 3    
plt.subplot(2,2,3)

sns.set_context("notebook")

sns.barplot(data=df_melted, x='Name of the States', y='Road Length', hue = 'Road Type')

plt.xticks(rotation=90)
plt.yscale('log')
plt.ylabel('Road Length - Log Scale')
plt.title('Screen Display View')
plt.legend( fontsize = '10')


plt.suptitle('Distribution of National, State and District roads')
plt.tight_layout()
plt.show()
```
